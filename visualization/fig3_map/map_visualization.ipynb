```python
import pandas as pd
import numpy as np
import re
import plotly.graph_objects as go
import plotly.express as px
```

df = pd.read_excel('cherry_blossom_1.xlsx')
df

data=[]
for i in range(0,df.shape[0]):
    state=df.iloc[i,11]
    for j in range(0,11):
        year=str(j+2011)
        if df.iloc[i,j]!='3/1':
            month=str(year)+'/'+df.iloc[i,j]
            actual=df.iloc[i,j]
        else:
            continue
        data.append([state,year,month,actual])
data= pd.DataFrame(data, columns=['state','year','month','actual'])
data

data.to_excel('map.xlsx')

```python
for i in range(0,data_new.shape[0]):
    data_new.iloc[i,1]=str(data_new.iloc[i,1])
data_new 
```

```python
data_new=data_new.sort_values(['year', 'actual'], ascending=[True, False])
data_new
```

```python
len(px.colors.sequential.Plotly3)
```

```python
fig = px.choropleth(data_new, locations='state', color='Viewing Date',
                           color_continuous_scale=px.colors.sequential.Purpor,
                           hover_name="state",
                           scope="usa",
                           animation_frame="year",
                    locationmode = "USA-states"
                           ##labels={'unemp':'unemployment rate'}
                          )
fig.add_scattergeo(
    locations=data_new['state'],
    locationmode='USA-states',
    text=data_new['state'],
    mode='text')
fig.update_layout(
    title={'text':'Map with the starting cherry blossom viewing date over 11 years',
           'xanchor':'center',
           'yanchor':'top',
           'x':0.5,
          'y':0.99})
fig.update_layout(
    title_font_family="Times New Roman",
    title_font_size=27.5)
fig.update_layout(
    legend_font_family="Times New Roman",
    legend_font_size=10,
    legend_grouptitlefont_family="Times New Roman"
)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)',plot_bgcolor='rgba(0,0,0,0)')
fig.update_layout( font_family="Times New Roman"
)
fig.write_html("map.html")
fig.show()
```
